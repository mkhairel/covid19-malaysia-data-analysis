```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully")
```

```python
import pandas as pd

df = pd.read_csv('cases_malaysia.csv')

df.head()
```

```python
df.info()
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['month'] = df['date'].dt.to_period('M')
```

```python
df['cases_cumulative'] = df['cases_new'].cumsum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df['date'], df['cases_new'])
plt.title('Daily New COVID-19 Cases in Malaysia')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.show()

print("""Insight:
Daily COVID-19 cases in Malaysia show multiple waves with significant spikes during certain periods, indicating different phases of the pandemic.""")
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['date'], df['cases_cumulative'])
plt.title('Cumulative COVID-19 Cases in Malaysia')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.show()
print("""Insight:
The cumulative case curve shows a steep increase over time, reflecting sustained transmission across multiple waves.""")
```

```python
top_10_days = df.sort_values(by='cases_new', ascending=False).head(10)
top_10_days[['date', 'cases_new']]

print ("""Insight:
The top 10 highest daily cases occurred during peak pandemic waves, highlighting periods of extreme healthcare pressure.""")
```

```python
monthly_cases = df.groupby('month')['cases_new'].sum()

monthly_cases.plot(kind='bar', figsize=(12,5), title='Monthly COVID-19 Cases in Malaysia')
plt.xlabel('Month')
plt.ylabel('Total Cases')
plt.show()
print("""Insight:
Monthly analysis reveals that certain months experienced significantly higher case volumes, suggesting seasonal or policy-related effects.""")
```

```python
import plotly.express as px
```

```python
fig = px.line(
    df,
    x='date',
    y='cases_new',
    title='Daily New COVID-19 Cases in Malaysia',
    labels={'cases_new':'New Cases', 'date':'Date'}
)

fig.show()
```

```python
monthly_df = df.groupby('month')['cases_new'].sum().reset_index()
monthly_df['month'] = monthly_df['month'].astype(str)

fig2 = px.bar(
    monthly_df,
    x='month',
    y='cases_new',
    title='Monthly COVID-19 Cases in Malaysia',
    labels={'cases_new':'Total Cases', 'month':'Month'}
)

fig2.show()
print ("""Daily cases: \"Pandemic waves clearly visible, peaks indicate critical periods\"\n\nMonthly cases: \"Some months had higher cumulative cases, possibly due to policy or seasonal effects\"\n\nTop 10 days: \"Peak days highlight healthcare system pressure\"""")
```
